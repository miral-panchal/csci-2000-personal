```python
import numpy as np
import matplotlib.pylab as plt

#Function to read string and store the outcomes into a list
def store(line1):
    var = []
    for i in line1.split(' '):
        #converts to float
        i = float(i)
        var.append([i])
    return var

#Opens file xy_points.dat
file = open('xy_points.dat', 'r') an 
#Reads first line
first_line = file.readline()
#Reads second line
second_line = file.readline()

#saves the result of the function called and stores it as an array
x = np.asarray(store(first_line))
y = np.asarray(store(second_line))

#gets the x values for when x is less than -2
x_final =x[x < -2]
#gets the y values for when x is less than -2
y_final= y[x < -2]

#gets the x_final values for when y_final is greater than or equal to -5 
x_final = x_final[y_final  >= -5]
#gets the y_final values for when y_final is greater than or equal to -5
y_final = y_final [y_final  >= -5]

#gets the x_final values for when y_final is less than or equal to 0
x_final = x_final[y_final  <= 0]
#gets the y_final values for when y_final is less than or equal to 0
y_final = y_final [y_final  <= 0]

r = []
#saves the range of every number into an array
for i in range (0,1000):
    r.append(np.sqrt((x[i]**2)+(y[i]-5)**2))

#converts r to an array
r = np.asarray(r)

#gets the x values for when the range is less than 5
x2_final = x[r <= 5]
#gets the y values for when the range is less than 5
y2_final = y[r <= 5]


#Sets x and y range and title for the graph
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title ('Higlighted points (100569725)')

#Prints graph for part b
plt.scatter(x, y,c='red', alpha=1)
#Prints graph for part c
plt.scatter(x_final , y_final  ,c='blue', alpha=1)
#Prints graph for part d
plt.scatter(x2_final , y2_final  ,c='green', alpha=1)

plt.show()
```
