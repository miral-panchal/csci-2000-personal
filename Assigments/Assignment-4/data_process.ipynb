```python
%pylab inline

from struct import *
import matplotlib.pyplot as plt
import numpy as np

with open('/home/miral/Desktop/CSCI-2000U-personal/Large_Data/halo.dat', 'rb') as file:
    tempN = file.readline()
    tempT = file.readline()
    tempData = file.read()
    
    x = []
    y = []
    z = []
    
    halo = []
    
    N_halo = unpack('i' *35, tempN[:-2])
    T_halo = unpack('d' *31, tempT[:-3])
    data_halo = unpack('i' * ((len(tempData) -3) // 4), tempData[:-3])
    
    i = 0
    for j in data_halo:
        if(i == 401661):
            break
        elif(i%3 == 0):
            x.append(j)
            i = i + 1
        elif(i%3 == 1):
            y.append(j)
            i = i + 1
        elif(i%3 == 2):
            z.append(j)
            i = i + 1
    k = 0    
    while k < (i/3):
        halo.append(math.sqrt((x[k]**2) + (y[k]**2) + (z[k]**2)))
        k = k + 1

hist, bins = np.histogram(halo, bins = 50)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2

plt.figure(figsize=(10, 10))
plt.bar(center, hist, width=width, color= 'red')
plt.ylabel('Velocity')
plt.title('Velocity histogram (halo) 100569725')
plt.show()


with open('/home/miral/Desktop/CSCI-2000U-personal/Large_Data/snapshot_010.dat', 'rb') as file:
    tempN = file.readline()
    tempT = file.readline()
    tempData = file.read()
    
    x = []
    y = []
    z = []
    
    halo = []
    
    
    N_010 = len(tempN)
    T_010 = len(tempT)
    data_010 = len(tempData)
    
    unpacked_N = unpack('i' *(math.floor(N_010/4)), tempN[:(-1*(N_010%4))])
    unpacked_T = unpack('d' *(math.floor(T_010/8)), tempT[:(-1*(T_010%8))])
    unpacked_data = unpack('i' * (math.floor((data_010)/ 4)), tempData[:(-1*(data_010%4))])

    i = 0
    for j in unpacked_data:
        if(i == 14679642):
            break
        elif(i%3 == 0):
            x.append(j)
            i = i + 1
        elif(i%3 == 1):
            y.append(j)
            i = i + 1
        elif(i%3 == 2):
            z.append(j)
            i = i + 1
    k = 0    
    while k < (i/3):
        halo.append(math.sqrt((x[k]**2) + (y[k]**2) + (z[k]**2)))
        k = k + 1

hist, bins = np.histogram(halo, bins = 50)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2

plt.figure(figsize=(10, 10))
plt.bar(center, hist, width=width, color= 'red')
plt.ylabel('Velocity')
plt.title('Velocity histogram (010) 100569725')
plt.show()
```
