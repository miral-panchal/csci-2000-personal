```python
#Miral Panchal - 100569725
%pylab inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from JSAnimation.IPython_display import display_animation

from mpl_toolkits.basemap import Basemap

import csv

def graph(title, clr, category): 
    Category = []
    Lat = []
    Long = []
    Cases = []
    
    #Size of the plot
    plt.figure(figsize=(10, 10))

    #Map of Earth
    earth = Basemap(projection='mill')
    earth.drawcoastlines(color='0.50', linewidth=0.5)
    earth.fillcontinents(color='1.0')

    #Opens file Source_data_for_CFR_vaccine_map_abridged.csv
    with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
        #Reads csv file
        reader = csv.DictReader(csvfile)
        for row in reader:
            #Adds the category to 'Category' list
            Category.append(row['Category'])
            #Adds the Latitude to 'Lat' list
            Lat.append((float)(row['Lat']))
            #Adds the long to 'Long' list
            Long.append((float)(row['Long']))
            #Adds the cases to 'Cases' list
            Cases.append((float)(row['Cases']))

    
    list(map(float, Lat))
    list(map(float, Long))
    list(map(float, Cases))

    #Changes the sizes of the marker based on the number of cases 
    for i in range (0, len(Category)):
        if (Category[i] == category):
            x, y = earth(Long[i], Lat[i])
        
            if (Cases[i] >= 0 and Cases[i] < 10000):
                earth.plot(x, y, clr,markersize=5,alpha=.75)
            
            elif (Cases[i] >= 10000 and Cases[i] < 20000):
                earth.plot(x, y, clr,markersize=10,alpha=.75)
            
            elif (Cases[i] >= 20000 and Cases[i] < 30000):
                earth.plot(x, y, clr,markersize=15,alpha=.75)
            
            elif (Cases[i] >= 30000 and Cases[i] < 40000):
                earth.plot(x, y, clr,markersize=20,alpha=.75)
       
            elif (Cases[i] >= 40000 and Cases[i] < 50000):
                earth.plot(x, y, clr,markersize=25,alpha=.75)
        
            elif (Cases[i] >= 50000 and Cases[i] < 60000):
                earth.plot(x, y, clr,markersize=30,alpha=.75)
       
            elif (Cases[i] >= 60000 and Cases[i] < 70000):
                earth.plot(x, y, clr,markersize=35,alpha=.75)
          
            elif (Cases[i] >= 70000 and Cases[i] < 80000):
                earth.plot(x, y, clr,markersize=40,alpha= .75)
            
            elif (Cases[i] >= 75000 and Cases[i] < 80000):
                earth.plot(x, y, clr,markersize=45,alpha=.75)
            
            elif (Cases[i] >= 80000 and Cases[i] < 90000):
                earth.plot(x, y, clr,markersize=50,alpha=.75)
            
            elif (Cases[i] >= 90000 and Cases[i] < 100000):
                earth.plot(x, y, clr,markersize=55,alpha=.75)
            
            elif (Cases[i] >= 100000 and Cases[i] < 110000):
                earth.plot(x, y, clr,markersize=60,alpha=.75)
            
            elif (Cases[i] >= 110000 and Cases[i] < 120000):
                earth.plot(x, y, clr,markersize=65,alpha=.75)
            
            elif (Cases[i] >= 120000 and Cases[i] < 130000):
                earth.plot(x, y, clr,markersize=70,alpha=.75)
            
            elif (Cases[i] >= 130000 and Cases[i] < 140000):
                earth.plot(x, y, clr,markersize=75,alpha=.75)
            

    
    plt.title(title)
    plt.show()

    
#Graph for Measles
graph('Measles Outbreak (100569725)','ro','Measles')
#Graph for Rubella
graph('Rubella Outbreak (100569725)','go','Rubella')
#Graph for Mumps
graph('Mumps Outbreak (100569725)','bo','Mumps')
#Graph for Polio
graph('Polio Outbreak (100569725)','yo','Polio')
```
