```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
import csv

a1 = []
a2 = []
a3 = []
a4 = []
a5 = []
a6 = []

with open('ISSN_D_tot.csv') as csvfile:
        reader = csv.DictReader(csvfile)        
        for row in reader:
            
            a1.append((int)(row['Year']))
            a2.append((int)(row['Month']))
            a3.append((int)(row['Day']))
            a4.append((float)(row['Date']))
            a5.append((int)(row['Number']))
            a6.append((int)(row['Flags']))
            
array = np.array([a1,a2,a3,a4,a5,a6])


print('Most recent observation was on: ',(int)(array[0][-1]),'-',(int)(array[1][-1]),'-',(int)(array[1][-1]))
print('Statistics of daily sunspot observations (',(int)(array[0][0]),'-present)')

mean = np.mean(array[4])
median = np.median(array[4])
deviation = np.std(array[4])

print('Mean: ',mean)
print('Median: ',median)
print('Standard Deviation: ',deviation)


plt.figure(figsize=(10, 10))
plt.scatter(array[4],array[3] ,c='cyan', alpha=1)
```
